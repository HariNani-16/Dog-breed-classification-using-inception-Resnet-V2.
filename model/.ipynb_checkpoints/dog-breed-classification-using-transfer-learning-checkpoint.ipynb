# About the Dataset
- This dataset provided with a training set and a test set of images of dogs. Each image has a filename that is its unique id. The dataset comprises 120 breeds of dogs. The goal of the competition is to create a classifier capable of determining a dog's breed from a photo

```python
import keras
import pandas as pd
import numpy as np
from keras.preprocessing.image import ImageDataGenerator
from keras.layers import Dense, Flatten, Activation, Dropout, Conv2D, MaxPooling2D
from keras.models import Sequential
from sklearn.metrics import classification_report, confusion_matrix
import cv2
import matplotlib.pyplot as plt
import os
from keras.applications import Xception
from keras.models import Sequential
from keras.callbacks import ModelCheckpoint, EarlyStopping
from keras.layers import Dense, Conv2D, MaxPooling2D, GlobalAveragePooling2D
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```python
path_t = '/kaggle/input/person-images/dogImages/dogImages/train/'
path_te = '/kaggle/input/person-images/dogImages/dogImages/valid/'
```

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                  rotation_range=30,
                                  width_shift_range=0.3,
                                  height_shift_range=0.3,
                                  fill_mode='nearest')
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(path_t,
                                                   target_size = (299,299),
                                                   batch_size = 64,
                                                   class_mode = 'categorical')

test_generator = test_datagen.flow_from_directory(path_te,
                                                 target_size = (299,299),
                                                 batch_size = 64,
                                                 class_mode = 'categorical')
```

```python
model_full = Xception(weights='imagenet', include_top = True, input_shape=(299,299,3))
print(model_full.summary())
```

```python
model = Xception(weights='imagenet', include_top = False, input_shape=(299,299,3))
print(model.summary())
```

```python
for layer in model.layers:
    layer.trainable = False
    #print(layer.trainable)
```

```python
xception_net = Sequential()
xception_net.add(model)
xception_net.add(GlobalAveragePooling2D())
xception_net.add(Dense(133, activation = 'softmax'))
xception_net.summary()
```

```python
xception_net.compile(optimizer= 'adam' , loss= 'categorical_crossentropy', metrics=['accuracy'])
```

```python
checkpoint = ModelCheckpoint('/kaggle/working/model.h5',
                             monitor="val_loss",
                             mode="min",
                             save_best_only = True,
                             verbose=1)

earlystop = EarlyStopping(monitor = 'val_loss', 
                          min_delta = 0, 
                          patience = 10,
                          verbose = 1,
                          restore_best_weights = True)

# we put our call backs into a callback list
callbacks = [earlystop, checkpoint]
```

```python
history = xception_net.fit(train_generator,
                               steps_per_epoch = 104,
                               epochs = 10,
                               verbose = 1,
                               validation_data = test_generator,
                               validation_steps = 13,
                               callbacks = callbacks
                               
                              )
```

```python
score = xception_net.evaluate_generator(test_generator, 13)
print('Test Loss', score[0])
print('Test Accuracy', score[1])
```

```python
history_dict = history.history

loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']
epochs = range(1, len(loss_values) + 1)

line1 = plt.plot(epochs, val_loss_values, label='Validation/Test Loss')
line2 = plt.plot(epochs, loss_values, label='Training Loss')
plt.setp(line1, linewidth=2.0, marker = '+', markersize=10.0)
plt.setp(line2, linewidth=2.0, marker = '4', markersize=10.0)
plt.xlabel('Epochs') 
plt.ylabel('Loss')
plt.grid(True)
plt.legend()
plt.show()
```

```python
history_dict = history.history

acc_values = history_dict['accuracy']
val_acc_values = history_dict['val_accuracy']
epochs = range(1, len(loss_values) + 1)

line1 = plt.plot(epochs, val_acc_values, label='Validation/Test Accuracy')
line2 = plt.plot(epochs, acc_values, label='Training Accuracy')
plt.setp(line1, linewidth=2.0, marker = '+', markersize=10.0)
plt.setp(line2, linewidth=2.0, marker = '4', markersize=10.0)
plt.xlabel('Epochs') 
plt.ylabel('Accuracy')
plt.grid(True)
plt.legend()
plt.show()
```
